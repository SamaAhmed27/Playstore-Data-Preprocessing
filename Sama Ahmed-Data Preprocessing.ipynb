# Data Preprocessing Project

### Let's start by reading and exploring the available data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set(rc={'figure.figsize': [10, 10]}, font_scale=1.3)
```

```python
df_ps = pd.read_csv('googleplaystore.csv')
df_ps
```

```python
df_ps = df_ps.reset_index()
df_ps
```

```python
df_ps.index = df_ps['index']
del df_ps['index']
df_ps
```

```python
df_ps.describe()
```

### Let's start data cleaning

Gather info, remove or replace nulls, and explore

```python
df_ps.info()
```

```python
# Let's remove the dollar sign from price
df_ps['Price'] = df_ps['Price'].apply(lambda x: x.split('$')[-1])
```

```python
# We shall convert these columns to be numerical
numerical_colms = ['Price', 'Reviews']

for col in numerical_colms:
    df_ps[col] = pd.to_numeric(df_ps[col], errors='coerce')

df_ps.info()
df_ps.describe()
```

From the above stats, it shows that the price category has an outlier of 400

Now, let's find null values

```python
df_ps.isnull().sum()
```

```python
# Let's drop rows that have few null values
df_ps.dropna(subset=['Type', 'Reviews', 'Price','Content Rating','Current Ver','Android Ver'], inplace=True)
df_ps.isnull().sum()
```

```python
# Fill missing values of ratings with median
df_ps['Rating'].fillna(df_ps['Rating'].median(), inplace=True)
```

```python
# Check again for null values
df_ps.isnull().sum()
```

### Fixing 'Ratings' Column

```python
# For ratings, there are values > 5, which is not true, so we shall drop this row 
df_ps[df_ps['Rating'] == 19]
```

```python
df_ps = df_ps[df_ps['Rating'] <= 5]
df_ps.describe()
```

### Fixing 'Installs' Column

```python
# Checking Installs Column
# Removed the '+' sign and commas
df_ps['Installs'] = df_ps['Installs'].apply(lambda x: x.split('+')[0].replace(',', ''))
df_ps
```

```python
df_ps['Installs'] = pd.to_numeric(df_ps['Installs'], errors='coerce')
df_ps.info()
```

### Fixing 'Size' Column

```python
# Check non-numeric values
df_ps[df_ps['Size'] == 'Varies with device']
```

For the 'Varies with device' part of Size, we have 2 approaches:
1. Remove them (1694/10841)
2. Fill them with the median of sizes

Let's go for the first method

```python
# Removing the varies with device rows
df_ps = df_ps[df_ps['Size'] != 'Varies with device']
df_ps
```

```python
# Now, we shall remove the size k and M, with the values, returning sizes all as MB
def convert_size(size):
    if 'M' in size:
        return float(size.replace('M', ''))
    elif 'k' in size:
        return float(size.replace('k', '')) / 1024  # Convert K to M

df_ps['Size'] = df_ps['Size'].apply(convert_size)
```

```python
df_ps
df_ps.describe()
```

### Fixing 'Category' Column

```python
# We can remove the underscores in the categories
df_ps['Category'] = df_ps['Category'].str.replace('_', ' ')
df_ps
```

### Fixing 'Android Ver' Column

The column has the minimumm supported android ver 'and up', we can only focus on the minimum android ver needed

```python
# Removing 'and up'
df_ps['Android Ver'] = df_ps['Android Ver'].str.replace('and up', '').str.strip()
```

```python
df_ps
```

```python
# Check non-numeric values
df_ps[df_ps['Android Ver'] == 'Varies with device']
```

### Fixing 'Genres' Column

```python
# We can split the several categories into primary and sub categories
df_ps['Primary Genre'] = df_ps['Genres'].apply(lambda x: x.split(';')[0])
df_ps['Sub Genre'] = df_ps['Genres'].apply(lambda x: x.split(';')[1] if ';' in x else None)
df_ps
```

```python
del df_ps['Genres']
df_ps
```

# Fixing 'Last Update' Column

```python
# Changing to date format
df_ps['Last Updated'] = pd.to_datetime(df_ps['Last Updated'], errors='coerce')
df_ps
```

# Questions

1.What is the most expensive app on the Play Store?

```python
most_expensive_app = df_ps[df_ps['Price'] == df_ps['Price'].max()]
print(most_expensive_app[['App', 'Price']])
```

2.Which genre has the highest number of apps?



```python
genre_counts = df_ps['Primary Genre'].value_counts()  # Using value_counts to count each genre
top_genre = genre_counts.idxmax()
top_genre_count = genre_counts.max()
print(f"The primary genre with the highest number of apps is: {top_genre} with {top_genre_count} apps.")

subgenre_counts = df_ps['Sub Genre'].value_counts()
top_subgenre = subgenre_counts.idxmax()
top_subgenre_count = subgenre_counts.max()
print(f"The sub genre with the highest number of apps is: {top_subgenre} with {top_subgenre_count} apps.")
```

3.What is the average size of free vs. paid apps?

```python
# Group by 'Type' and calculate the average size
average_size = df_ps.groupby('Type')['Size'].mean()
print(average_size)
```

4.What are the top 5 most expensive apps with a perfect rating (5)?

```python
top_5_expensive_perfect_rating = df_ps[df_ps['Rating'] == 5].nlargest(5, 'Price')
print(top_5_expensive_perfect_rating[['App', 'Price', 'Rating']])
```

5.How many apps have received more than 50K reviews?

```python
apps_above_50k_reviews = df_ps[df_ps['Reviews'] > 50000]
num_apps_above_50k = apps_above_50k_reviews.shape[0] # Shape[0] -> no. of rows
print(f"Number of apps with more than 50K reviews: {num_apps_above_50k}")
```

6.What is the average price of apps, grouped by genre and number of installs?

```python
# Group by 'Genres' and 'Installs' then get average
#This way we can create a simple datafram which we use to obtain the info we need
average_price_by_genre_installs = df_ps.groupby(['Primary Genre', 'Installs'])['Price'].mean().reset_index() 
print(average_price_by_genre_installs)
```

7.How many apps have a rating higher than 4.7, and what is their average price?

```python
high_rating_apps = df_ps[df_ps['Rating'] > 4.7]
num_high_rating_apps = high_rating_apps.shape[0]
avg_price_high_rating_apps = high_rating_apps['Price'].mean()
print(f"Number of apps with a rating higher than 4.7: {num_high_rating_apps}")
print(f"Average price of apps with a rating higher than 4.7: {avg_price_high_rating_apps}")
```

8.What is Google's estimated revenue from apps with 5,000,000+ installs? (Assuming Google takes a 30% cut from app sales

```python
apps_above_5m_installs = df_ps[df_ps['Installs'] >= 5000000]
estimated_revenue = (apps_above_5m_installs['Price'] * 0.30).sum()
print(f"Estimated revenue from apps with 5,000,000+ installs: {estimated_revenue}")
```

9.What are the maximum and minimum sizes of free vs. paid apps?

```python
size_stats = df_ps.groupby('Type')['Size'].agg(['max', 'min'])
print(size_stats)
```

10.Is there a correlation between an app’s rating, number of reviews, size, and its price?

```python
# correlation matrix
correlation_matrix = df_ps[['Rating', 'Reviews', 'Size', 'Price']].corr()
print(correlation_matrix)
sns.heatmap(correlation_matrix) # Visualize using the heatmap
```

11.How many apps exist for each type (free/paid) across different content ratings?

```python
df_ps['Content Rating'].unique()
# To identify number of content ratings available
```

```python
apps_by_type_content_rating = df_ps.groupby(['Type', 'Content Rating']).size().reset_index(name='App Count')
print(apps_by_type_content_rating)
```

12.How many apps are compatible with Android version 4.x?

```python
android_4_apps = df_ps[df_ps['Android Ver'].str.startswith('4.', na=False)]
num_android_4_apps = android_4_apps.shape[0]
print(f"Number of apps compatible with Android version 4.x: {num_android_4_apps}")
```
